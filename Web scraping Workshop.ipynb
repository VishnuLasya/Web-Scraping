```python
URL='https://forecast.weather.gov/MapClick.php?lat=38.85279660000003&lon=-77.34201179999997#.YFS1LNqSlPY'
page=requests.get(URL)
page.status_code #if 200 its success in request the url
```

```python
soup=bs(page.content,'html')
soup #display all the data in thhe html
```

# all the lists

```python
time_period_data=[t.get_text() for t in soup.select(".tombstone-container .period-name ")]
time_period_data
```

```python
short_desc_data=[t.get_text() for t in soup.select(".tombstone-container .short-desc ")]
short_desc_data
```

```python
temperature_data=[t.get_text() for t in soup.select(".tombstone-container .temp ")]
temperature_data
```

```python
title_data=[t['title']for t in soup.select(".tombstone-container img ")]
title_data
```

```python
for t in soup.select(".tombstone-container .period-name"):
    print(t)
```

# lists into a table

```python
import pandas as pd
data_dict={"time_period_data":time_period_data,"short_desc_data":short_desc_data,"temperature_data":temperature_data,"title_data":title_data}
df=pd.DataFrame.from_dict(data_dict)
df
```

```python
#df.to_csv('D:/George Mason/web scraping/weather.csv',index=False)
weather_data= list(zip(time_period_data,short_desc_data,temperature_data,title_data))
weather_data

        
```

```python
with open('D:/George Mason/web scraping/weather.csv','w',encoding='utf-8') as file:
    writer=csv.writer(file,delimiter=',')
    for i in weather_data:
        l=list(i)
        writer.writerow(l)
        
#this automatically creates a file  weather.csv in the folder        
```
